```python
import matplotlib.pyplot as plt

# Count values and calculate percentages
attrition_counts = HRdata['Attrition'].value_counts()
total = attrition_counts.sum()
percentages = (attrition_counts / total) * 100

# Plot bar chart
plt.figure(figsize=(8, 6))
ax = attrition_counts.plot(kind='bar', color=['green', 'red'])

# Add title and labels
plt.title("Employee Attrition Distribution", fontsize=14)
plt.xlabel("Attrition", fontsize=12)
plt.ylabel("Number of Employees", fontsize=12)

# Add percentage labels on top of bars
for i, val in enumerate(attrition_counts):
    plt.text(i, val + 10, f"{percentages[i]:.1f}%", ha='center', fontsize=12, fontweight='bold')

# Show plot
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Load dataset
HRdata = pd.read_csv("C:/Users/Winquistt/DSC 680/Week 9 thru 12/HR-Employee-Attrition.csv")

# Convert 'Attrition' (target) to binary (Yes = 1, No = 0)
HRdata['Attrition'] = HRdata['Attrition'].map({'Yes': 1, 'No': 0})

# Select features (drop non-predictive columns like EmployeeNumber)
features = HRdata.drop(columns=['Attrition', 'EmployeeNumber'])

# Identify categorical columns and encode them
cat_cols = features.select_dtypes(include=['object']).columns

# Apply Label Encoding to categorical features
le = LabelEncoder()
for col in cat_cols:
    features[col] = le.fit_transform(features[col])

# Standardize numerical features
scaler = StandardScaler()
num_cols = features.select_dtypes(include=['int64', 'float64']).columns
features[num_cols] = scaler.fit_transform(features[num_cols])

# Define target variable
target = HRdata['Attrition']

# Split dataset into train (80%) and test (20%)
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)

# Train Random Forest Classifier
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train)

# Predict on test set
y_pred = rf_model.predict(X_test)

# Evaluate model performance
accuracy = accuracy_score(y_test, y_pred)
print(f"Model Accuracy: {accuracy:.2f}")

# Print classification report
print("\nClassification Report:\n", classification_report(y_test, y_pred))

# Confusion matrix visualization
plt.figure(figsize=(6, 4))
sns.heatmap(confusion_matrix(y_test, y_pred), annot=True, fmt='d', cmap="Greens", xticklabels=['No Attrition', 'Attrition'], yticklabels=['No Attrition', 'Attrition'])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()

# Feature Importance
importances = pd.Series(rf_model.feature_importances_, index=features.columns).sort_values(ascending=False)
plt.figure(figsize=(10, 6))
importances[:10].plot(kind='bar', color='teal')
plt.title("Top 10 Feature Importances")
plt.ylabel("Importance Score")
plt.show()
```

The model achieved 88% accuracy, which is quite strong!
 ** Accuracy: 88% (Model correctly predicts attrition 88% of the time).
**Precision (1 - Attrition cases): 80% (When the model predicts attrition, it is correct 80% of the time).
**Recall (1 - Attrition cases): 10% (The model only catches 10% of actual attrition cases).
**F1-score (Attrition): 18% (Low recall suggests the model struggles to detect attrition cases).
